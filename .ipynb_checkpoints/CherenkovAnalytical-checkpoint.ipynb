```python
import pandas as pd
import numpy as np
import ROOT as rt

rt.gStyle.SetOptStat(0)

planckConstant = 6.63e-34
speedOfLight = 3.0e8
electronCharge = 1.6e-19
```

```python
df = pd.read_csv('SapphireRefractiveIndex.csv')
photonEnergy = np.array(df['eV'])
refract = np.array(df['n'])

cRefractiveIndex = rt.TCanvas("cRefractiveIndex", "", 700, 1200)
cRefractiveIndex.Divide(1, 2)

cRefractiveIndex.cd(1)
gRefractiveIndex = rt.TGraph(len(df), photonEnergy, refract)
gRefractiveIndex.SetTitle(";photon energy [eV]; sapphire refractive index")
gRefractiveIndex.Draw('AC')

cRefractiveIndex.cd(2)
hRefractiveIndex = rt.TH1F('hRefract', ';energy [eV];refractive index', 30, 0., 6.5)
for i,energy in enumerate(photonEnergy):
    hRefractiveIndex.SetBinContent(hRefractiveIndex.FindBin(energy), refract[i])
    
hRefractiveIndex.Draw()
cRefractiveIndex.Draw()
```

```python
cRefractiveIndex = rt.TCanvas("cRefractiveIndex", "", 700, 600)

hRefractWl = rt.TH1F('hRefractWl', ';wavelength [nm];refractive index', 26, 200., 600.)
for i,energy in enumerate(photonEnergy):
    photonWl = planckConstant*speedOfLight/(energy*electronCharge)*1.e9
    hRefractWl.SetBinContent(hRefractWl.FindBin(photonWl), refract[i])

hRefractWl.Draw()
cRefractiveIndex.Draw()
```
